Here are some tips for colors:

Colors can be specified by single letters:

    k - black
    w - white
    c - cyan
    m - magenta

    r - red
    y - yellow

or by html color codes ('red', 'orange', 'steelblue', etc)

or by hex color codes (#BDA493, #63C638, etc - A very handy color picker can be found here)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

t = np.arange(0., 10., 0.5) # similar to how we use range function in python

# This is the first figure (Modification: Changed the color of the line)
plt.figure(1)
plt.plot(t, t, 'r--', label='linear', color = "steelblue")
plt.legend(loc='upper left', shadow=True, fontsize='medium')
plt.title('This is figure 1')

#Extra Feature: Using annotate to add annotation to a specific point on plot 
plt.annotate('Slope = 1', xy =(3.3, 1),
                xytext =(3, 1.8), )
plt.show()
```

```python
# This is the second figure Modification: Changed Figure Size and Resolution
plt.figure(2, figsize=(8,8), dpi=100)
plt.plot(t, t**2, 'bs-', label='square')

plt.legend(loc='upper left', shadow=True, fontsize='medium')
plt.title('This is figure 2')

#Extra Feature: Using twiny (adds a new axis that overlays the X-axis but is on top of the graph)
plt.twiny(ax=None)

plt.show()
```

```python
# This is the third figure
plt.figure(3)
plt.plot(t, t**3, 'g^-', label = 'cubic')

plt.legend(loc='upper left', shadow=True, fontsize='medium')
plt.title('This is figure 3')

#Labeling X-Axis and Y-Axis (Modification)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#Extra Feature: Modification using suptitle which adds a subtitle for the figure
plt.suptitle("Subtitle")

plt.show()
```

Can you create a graph that shows the ED admission rate for each county, plotting under 18 against 18 and over, displaying separately each sex?

```python
#Reading in csv as dataframe
df_asthma = pd.read_csv("../../../trgn515_2023/datasets/asthma.csv")

#Filtering for Strata Names with Male or Female
df1 = df_asthma[df_asthma['Strata Name'].isin(['Male', 'Female'])]

#Filtering to remove All Ages from Age Group Column
df2 = df1[df1["Age Group"].isin(["All Ages"]) == False]
```

```python
#Creating Pivot Table
df3_pivot = df2.pivot_table( values='Rate', 
                            columns=['Strata Name','Age Group'], 
                            index='Geography'
                          )
df3_pivot.plot(kind = 'bar', figsize =(25,25))

plt.legend(loc='upper left', shadow=True, fontsize='medium')
plt.ylabel('ED Admission Rate')
```
